В модели почти свободных электронов, которую предложили Крониг и Пенни, рассматривается движение электрона в линейной цепочке прямоугольных потенциальных ям. Ширина ям равна $a$, и они отдельны друг от друга потенциальными барьерами толщиной $b$ и высотой $V_0$ . Длина цепочки равна $L$, а период цепочки равен $c = a+b$.

![table](./tables/table4.gif)

Пусть $E$ – энергия электрона. Состояние электрона описывается уравнением Шредингера:

$$\frac{d^2\psi}{dx^2}+\frac{2m}{h^2}(E-U)\psi=0$$

Решение этого уравнение разбивается на две части:
1. Для области I, где потенциальная энергия $V = 0$, волновая функция, являющаяся решением уравнения, может быть представлена в виде:

$$\psi_1(x)\ =\ Ae^{iax}+\ Be^{-iax}$$

Первое слагаемое в уравнении соответствует прямой волне, а второе – волне, отраженной от потенциального барьера.

2. Для области II, где $V = V_0$, волновую функцию можно записать в виде

$$\psi_2(x)\ =\ Ce^{ \beta x}+\ De^{ -\beta x}$$

В уравнениях коэффициенты $a\ =\ \sqrt{\frac{2mE}{h^2}}$, $\beta\ =\ \sqrt{\frac{2{(U}_0-E)}{h^2}}$; величины A, B, C, D - некоторые константы.

Подставим вместо $\psi_1$ и $\psi_2$ одномерную функцию Блоха $\psi(x)=V(x)e^{ikx}$ и найдем вид периодической функции $V(x)$ в областях I и II:

$$V_1(x)=Ae^{(a-ik)x}+Be^{-(a+ik)x}$$

$$V_2(x)=Ce^{(\beta-ik)x}+De^{-(\beta+ik)x}$$

Для того чтобы определить коэффициенты $A, B, C, D$, используем тот факт, что функция $V(x)$ и ее первая производная непрерывны в местах скачка потенциала. Например, в точке $x=0$

$V_1(0) = V_2(0)$, ${(\frac{dV_1}{dx})}_{x=0}={(\frac{dV_2}{dx})}_{x=0}$

Кроме того, функции $V(x)$ обладают свойством периодичности с периодом $c = a + b$

$V_1(a) = V_2(-b)$, ${(\frac{dV_1}{dx})}_{x=a}={(\frac{dV_2}{dx})}_{x=-b}$

Подставляя эти условия в систему уравнений для потенциалов $V_1(x)$ и $V_2(x)$, получим четыре уравнения. Известно, что такая система уравнений будет иметь ненулевые решения при условии, что определитель, составленный из коэффициентов при $A, B, C, D,$ будет равен нулю. Это требование приводит к уравнению:

$$\frac{\beta^2-\alpha^2}{2\alpha\beta}sh(\beta b)sin(\alpha a)+ch(\beta b)cos(\alpha a)=cosk(a+b)$$

Рассмотрим предельный случай, устремив ширину барьера $b$ к нулю и увеличим высоту барьера, устремив ее к бесконечности. При этом потребуем, чтобы произведение $bV_0$ оставалось постоянным. Обозначим $bV_0=\frac{h^2}{ma}P,\ тогда\ P=\frac{ma}{h^2bV_0}$, где $P$ − величина, характеризующая прозрачность барьера. Для этого предельного случая уравнение примет вид $\frac{Psin(\alpha a)}{\alpha a}+cos(\alpha a)=cos(ka)$

Уравнение выражает зависимость энергии электрона, которая входит в коэффициент $α$, от волнового числа $k$ для барьеров различной прозрачности $P$. Поскольку $cos(ka),\ где -1≤cos(ka)≤1$, то и левая часть уравнения лежит в этих же пределах. Эти значения определяют области разрешенных энергий электрона – энергетические зоны. Они отделены друг от друга полосами запрещенных энергий – запрещенными зонами. С увеличением энергии электрона ширина разрешенных зон увеличивается, а запрещенных уменьшается. Ширина зон зависит также от параметра прозрачности барьера $P$. При $P\rightarrow\infty$ разрешенные зоны сужаются, превращаясь в дискретные уровни, соответствующие $\alpha a=\pi n,\ где\ n = ∓1,∓2,∓3,...$. Тем самым мы приходим к случаю электрона в изолированном атоме. При стремлении прозрачности барьера к нулю, наоборот, исчезают запрещенные зоны, и электрон становится свободным.

Внутри каждой зоны энергия электрона растет с ростом значения модуля волнового вектора. При значениях $k=n\frac{\pi}{a},\ где\ n = ∓1,∓2,∓3,... $, энергия терпит разрыв

```python
import numpy as np
import matplotlib.pyplot as plt

# Значения x для графика
values = np.arange(-4, 4, 0.001)

def f(alpha, a, P):
    return np.cos(alpha * a) + P * np.sin(alpha * a) / (alpha * a)

# Создание фигуры и осей
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15, 5))

# Значения P для каждого графика
P_values = [0, 5, 9999999999999]
labels = ['P=0', 'P=5', 'P=∞']

for i, ax in enumerate(axes):
    P = P_values[i]

    ax.set_xlabel('αa')
    ax.set_ylabel('f(α, a)')
    ax.set_title(labels[i])

    # Добавление пунктирных линий на y = 1 и y = -1
    ax.axhline(y=1, linestyle="--", c='black', linewidth=1)
    ax.axhline(y=-1, linestyle="--", c='black', linewidth=1)

    # Вычисление всех значений функции энергетических зон
    ax.plot(values * (-1 * values), f(values, values, P), color='C0', linewidth=2)
    # Для отрицательных значений
    ax.plot(values * (1 * values), f(values, values, P), color='C0', linewidth=2)
    # Для положительных значений

    # Построение областей разрешенных и запрещенных зон со штриховкой
    ax.fill_between(values * (-1 * values), -1, 1, where=np.abs(f(values, values, P)) < 1,
                    color='green', alpha=0.3, hatch='//')
    ax.fill_between(values * (1 * values), -1, 1, where=np.abs(f(values, values, P)) < 1,
                    color='green', alpha=0.3, hatch='//')

    # Построение границ разрешенных зон и запрещенных зон
    ax.plot(values * (-1 * values), [1 if abs(f(i, i, P)) < 1 else -1 for i in values],
            color='red', linewidth=1)
    ax.plot(values * (1 * values), [1 if abs(f(i, i, P)) < 1 else -1 for i in values],
            color='red', linewidth=1)

    # Вычисление процента запрещенных зон
    barsValues = [1 if abs(f(i, i, P)) < 1 else -1 for i in values]
    print(f"Процент запрещенных зон ({labels[i]})"
          f": {barsValues.count(-1) / len(barsValues) * 100 : .2f}" + "%")

# Отображение графиков
plt.tight_layout()
plt.show()
```

Таким образом, энергия электрона в периодическом поле не может принимать любое значение. Она ограничена рядом полос (зон) разрешенных значений, отделенных друг от друга запрещенными зонами - энергетический спектр электрона в периодическом поле имеет зонную структуру. Ширина разрешенных зон определяется степенью связанности электрона внутри потенциальной ямы.

В случае совершенно свободного электрна $(P = 0)$энергия электрона может принимать любые значения.

В случае совершенно непрозрачных стенок потенциальной ямы $(P = \infty)$разрешенные зоны выраждаются в дискретные энергетические уровни

```python
P_values = np.linspace(0, 15, 100)

percentages = []
for P in P_values:
    values = np.arange(-4, 4, 0.001)
    barsValues = [1 if abs(f(i, i, P)) < 1 else -1 for i in values]
    percentage = barsValues.count(-1) / len(barsValues) * 100
    percentages.append(percentage)

plt.plot(P_values, percentages)
plt.xlabel('Parameter P')
plt.ylabel('Percentage of Forbidden Zones')
plt.grid(True)
plt.show()
```

Были построены модели и графики с помощью языка Python. Выводы, полученные из работы:

• Свободный электрон: Если электрон совершенно свободен и не подвергается воздействию потенциала, то зонная структура будет обладать непрерывным спектром энергий. Запрещенных зон не будет, и энергия электрона может принимать любые значения в непрерывном диапазоне.

• Электрон внутри потенциальной ямы: Если электрон заперт внутри одной потенциальной ямы, где стенки ямы являются непроницаемыми для электронов, то зонная структура будет иметь дискретный спектр энергий. Появятся запрещенные зоны, где энергия электрона будет иметь ограниченные значения в пределах ямы, и между запрещенными зонами будут находиться разрешенные зоны, где энергия может принимать любые значения.

• Промежуточные случаи: В промежуточных случаях, когда электрон находится внутри кристалла, но не полностью заперт внутри одной ямы, зонная структура будет содержать как запрещенные, так и разрешенные зоны. Ширина запрещенных зон будет зависеть от конфигурации потенциала в кристалле и расположения электрона относительно потенциальных ям.

• Ширина запрещенных зон зависит от конфигурации потенциала в кристалле.

В целом, модель Кронига-Пени позволяет анализировать влияние потенциальных энергий на зонную структуру одномерного кристалла и предсказывать изменения ширины запрещенной зоны в различных условиях. Это позволяет лучше понять поведение электронов в полупроводниковых и других кристаллических материалах и их взаимодействие с потенциальными барьерами и ямами.
