Используя уравнение Шредингера, найти связные состояния и соответствующие им собственные значения в случае прямоугольной потенциальной ямы:

$$V(x)\ = \ 0,|x|>a;-V_0,-a≤x≤a$$

В области $x<−a$, где потенциальное поле равно нулю, стационарное уравнение Шрeдингера запишется как:

$$\frac{d^2\psi}{dx^2}=k^2\psi$$

Где: $$k\ =\ \frac{\sqrt{-2mE}}{h}$$

Величина $k$ вещественна и положительна. Общим решением уравнения является:

$$\psi\ =\ {\rm Ae}^{-kx}\ +{\rm Be}^{\ kx}$$

при x → −∞ первое слагаемое стремится к бесконечности, следовательно $A = 0$ и $$\psi(x)\ =\ {\rm Be}^{kx}\ ,для\ x < -a$$

В области $−a < x < a$, где $V(x) = −V_0$, уравнение Шрeдингера будет выглядеть:

$$\frac{d^2\psi}{dx^2}={-l}^2\psi$$

Где $$l\ =\ \frac{\sqrt{2m(E+V_0)}}{h}$$

$l$ – вещественно и положительно, $а −U_0 < E < 0$. Общее решение уравнения:

$$\psi(x)\ =\ C\ sin(lx)\ +\ D\ cos(lx),\ при - a < x < a$$

В области $x < −a$ решение аналогично области $x < −a$, но в ноль при $x → ∞$ обращается второе слагаемое. Для нее $$\psi(x)\ ={\rm Fe}^{-kx},\ для\ x > a$$

Потенциал инвариантен по отношению к операции инверсии. Поэтому решения уравнения Шрeдингера должны быть либо четными, либо нечетными функциями x. Четным решениям отвечает уравнение  при $C = 0$, нечетным – при $D = 0$ (при четных решениях внутри ямы волновые функции в основном сохраняют косинусоидальный характер, при нечетных – синусоидальный, а за пределами убывают по экспоненте). Рассмотрим только четный случай, ищем решение функции:

$$\psi\ =\ Fe-kx,\ при\  x > a;\ D cos(lx),\ при\  - a < x < a;\ ψ(-x),при\  x < 0$$

Из условия непрерывности $ψ$ в точке $x = a$ следует $${\rm Fe}^{-ka}=Dcos\left(la\right)$$

Из условия непрерывности производной $dψ/dx$ следует $${-kFe}^{-ka}\ =\ -lD\ sin(la)$$

Поделив уравнения друг на друга, получим $$\frac{k}{l}\ =\ tg(la)$$

Это формула для разрешенных уровней энергии, поскольку $k$ и $l$ являются функциями от $E$. Введем новые обозначения

$$z=la,\ и\ z_0\ =\ \frac{a}{h}\sqrt{2mV_0}$$

Поскольку $(k^2+l^2)\ =\ \frac{2mV_0}{h^2}$, то в новых обозначениях $ka=\sqrt{{z_0}^2-z^2}$, и уравнение будет

$$tg(z)\ =\ \sqrt{{(\frac{z_0}{z})}^2-1}$$

С учетом выражения для параметра $z_0$, можно сделать вывод, что число уровней пропорционально ширине ямы $a$ и корню глубины ямы $V_0$


$$tg(la)\ =\ \sqrt{\frac{2mV_0}{h^2l^2}-1}$$

Так как $$arctg(x)\ =arcsin(\frac{x}{\sqrt{x^2+1}})$$

$$la\ =\ arcsin(\frac{\sqrt{\frac{2mV_0}{h^2l^2}-1}}{\sqrt{\frac{2mV_0}{h^2l^2}}})\ +\pi n$$

$$la\ =\ arcsin(\sqrt{1-\frac{h^2l^2}{2mV_0}})+\pi n$$

Условие для существования корней:

$$1=\frac{h^2l^2}{2mV_0}$$

$$l_{max}=\frac{\sqrt{2mV_0}}{h}$$

Импортируем необходимые библиотеки:

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import constants
```

```python
def f(U, x, a):
    if (abs(x) < a):
        return -U
    else:
        return 0
```

```python
U = 1.6e-17
a = 2e-10
m = 9.1e-31
h = constants.hbar
```

```python
x = np.linspace(-2*a, 2*a, 1000)

y = []
for i in range(len(x)):
    y.append(f(U, x[i], a))


# Рассчитываем lmax
lmax = np.sqrt(2 * m * U) / h

# Создаем массив значений l от 0 до lmax
l_values = np.linspace(0, lmax, 1000)

# Рассчитываем количество полос пересечения
periods = np.ceil(lmax * a / np.pi)


# Инициализируем массив для точек пересечения
intersection_points_x = []
intersection_points_y = []

rightFunctionValues = []

leftFunctionValues = a * l_values

for i in range(int(periods)):
    rightFunction = i * constants.pi + np.arcsin(np.sqrt(abs(1 - ((h ** 2 * l_values ** 2) / (2 * m * U)))))

    indices = np.where(np.isclose(rightFunction, leftFunctionValues, atol=0.1))

    # Находим точки пересечения
    for index in indices:
        mean_value = np.mean(rightFunction[index])
        rounded_mean = round(mean_value, 5)
        intersection_points_x.append(rounded_mean / a)
        intersection_points_y.append(rounded_mean)

    plt.plot(l_values, rightFunction)


# Построение графика
plt.plot(intersection_points_x, intersection_points_y, 'ro')
plt.plot(l_values, a * l_values)
plt.xlabel('l')
plt.ylabel('Function Value')
plt.grid(True)
plt.show()
```

 Как видно из графика, число точек пересечения растет с ростом параметра $l$
 При малых значениях параметра имеется по крайней мере одно пересечение графиков для четных решений. Это означает, что в любой яме имеется по крайней мере одно связанное четное состояние.

Собственные значения находятся как абсциссы точек пересечения кривых с прямой. Упомянутые кривые, разумеется, зависят от параметра $z_0$, определяемого размерами ямы.

С увеличением размеров ямы ее вместимость возрастает: число связанных состояний растет линейно с ростом $z_0$, образуя чередующиеся серии состояний с положительной четностью. Что касается собственных функций, то они следуют общему правилу: чем больше у них нулей, тем выше их положение на шкале энергий.


```python
# Вычислим энергии из найденных значений l и отметим их на графике
energies = [0] * len(intersection_points_x)

plt.plot(x, y)

# Определим пси-функцию
def psi(x, l):
    # Коэффициент, чтобы функции красиво выглядели
    coefficient = 1e18
    return np.cos(l*x / 2)/coefficient

for index in range(len(intersection_points_x)):
    energies[index] = (h ** 2) * intersection_points_x[index] ** 2 / (2 * m)
    print(f"E{index} {energies[index]}")
    plt.axhline(y=energies[index] - U, linestyle="--", c='red', linewidth=0.6)

    # Нарисуем на графике графики волновых функций
    yPsi = []

    for xIndex in range(len(x)):
            yPsi.append(psi(x[xIndex], intersection_points_x[index]) + (energies[index] - U))
    plt.plot(x, yPsi, c='green')

plt.xlabel("a", fontsize=14)
plt.ylabel("U", rotation=0, fontsize=14)
plt.grid()
plt.show()
```

Найти также собственные функции и собственные значения для осцилляторного потенциала $V(x)=\frac{1}{2}mw^2x^2$. Построить графически собственные функции. Рассмотреть случай, когда в точке $x = 0$ вводится бесконечно узкая и бесконечная полупроницаемая перегородка. Выявить влияние такой перегородки на стационарные состояния.

В классической физике гармоническим осциллятором называют частицу, на которую действует сила, прямо пропорциональная величине отклонения частицы от положения равновесия и направленная к нему. В классической механике возможно решение задачи о гармоническом осцилляторе на основе второго закона Ньютона (динамическое описание). В квантовой механике понятие силы не используется, и информацию о состояниях гармонического осциллятора получают в результате решения уравнения Шрѐдингера, в котором оператор потенциальной энергии имеет вид, соответствующий поставленной задаче.


Рассмотрим пример с частицей массой m, прикрепленную к пружинке. По закону Гука на такую частицу действует сила $F = −kx$. Потенциал такой частицы опишется как

$$V(x)\ =\ -\ \int_{-\infty}^{\infty}{(-kx)dx\ =\ V_0\ +0.5kx^2}$$

Таким потенциалом можно аппроксимировать произвольный потенциал в окрестности устойчивого положения равновесия. Если $V(x) = 0$, то гамильтониан системы будет

$$\hat{H}=\widehat{E_k}+\hat{U}=-\frac{h^2}{2m}*\frac{d^2}{dx^2}+0.5kx^2$$

Одномерное стационарное уравнение Шрeдингера для выбранного  потенциала

$$-\frac{h^2}{2m}*\frac{d^2\psi(x)}{dx^2}+0.5kx^2\psi(x)\ =\ E\psi(x)$$

Решения такого уравнения

$${\psi(x)}_{n\ }=\ A_nH_n(\beta x)e^{-\beta x^2/2},\ n\ =\ 0,1,2,...$$

где $H_n$ – полином Эрмита степени $n$

$$H_n(\varepsilon)={(-1)}^ne^{\varepsilon^2}\frac{d^n}{d\varepsilon^n}(e^{-\varepsilon^2})$$

а величины $β$ и $A_n$ обозначают

$$\beta^2=\frac{wm}{h},\ A_n\ =\ \frac{1}{\sqrt{2^nn!}}{(\frac{mw}{\pi h})}^{1/4},w^2=k/m$$

Уровни энергии такой системы

$$E_n=hw(n+0.5)$$

Кроме того, из выражения для энергии следует, что при $n = 0$ осциллятор имеет отличную от нуля минимальную (так называемую «нулевую») энергию: $$E_0 \neq 0.5hw;$$ то есть осциллятор не может находиться на дне ямы. Можно показать, что величина $E_0$ – наименьшее значение энергии, совместимое с соотношением неопределенностей.

Заметим, что уровни находятся на одинаковом расстоянии друг от друга.

```python
# Построим график осцилляторного потенциала
def pot(omega, X):
    return 0.5 * constants.electron_mass * pow(omega, 2) * pow(X, 2)

x = np.arange(-3e-2, 3e-2, 1e-4)
y = []
omega = 1
alpha = constants.electron_mass * omega / constants.hbar

for index in range(len(x)):
    y.append(pot(omega, x[index]))

plt.plot(x, y, linewidth=3)

# Определим функцию для Эрмитова многочлена
def hermite(n, x):
    array = [0] * (n) + [1]
    hermite = np.polynomial.hermite.Hermite(array)
    return hermite(x * np.sqrt(alpha))


# График собственных функций осциллографа
def oscillographFunctions(n, x, E):
    points = []
    for index in range(len(x)):
        norm = 6e-36  # Коэффициент нормализации,
        #чтобы графики получились не огромные
        coefficient = 1 / (np.sqrt(pow(2, n) * math.factorial(n))) \
        * pow((alpha / constants.pi), 0.25)
        point = norm * coefficient * np.exp(-0.5 * alpha * pow(x[index], 2))\
        * hermite(n, x[index]) + E
        # + E чтобы поднять график на определенную высоту
        points.append(point)

    return points


# Нарисуем линии энергетических уровней
# Собственные значения — значения энергий
ownValues = []

n = 0
while True:
    E = (n + 0.5) * constants.hbar * omega

    if (E > y[0]):
        break

    ownValues.append(E)
    print(f"E {E}")
    plt.plot(x, [E] * len(x), linewidth=0.6, c='red')
    plt.plot(x, oscillographFunctions(n, x, E), c='green', linewidth=1)
    n += 1

plt.xlabel("x", fontsize=14)
plt.ylabel("Ψ", rotation=0, fontsize=14)
plt.grid()
plt.show()
```

Рассмотрим случай введения бесконечно-узкой и бесконечно-полупроницаемой перегородки в точке $x=0$. Полупроницаемую перегородку, разделяющую всю область на две равные части, можно получить как предельный случай барьера конечной ширины $2ε$ (между точками $x=-ε$ и $x=ε$) и конечной высоты $U$. Для краткости введем обозначения:

$$k^2=\frac{2mE}{h^2}\ и\ \frac{2mE}{h^2}(U-E)\ =\ \aleph^2$$

Кроме двух граничных условий $u(±a) = 0$ благодаря наличию барьера мы имеем еще четыре граничных условия, так как функции $u(x)$ и $u ′ (x)$ должны быть непрерывны в точках $x = ±\varepsilon$. Удовлетворяя первым двум условиям и беря решение в действительной форме, имеем

$$u\ =\ A_1sink(x+a),-a≤x≤ε;\ Be-ℵx,-ε ≤x ≤ε;\ A_2sink(x-a),ε≤x≤a$$

$$u\ =\ A_1sink(x+a),-a≤x≤ε;\ Be-ℵx,-ε ≤x ≤ε;\ A_2sink(x-a),ε≤x≤a$$

Требование непрерывности при $x = ±ε$ теперь дает:

$$u(-\varepsilon)\ =A_1sink(a\ -\ \varepsilon)=Be^{m\varepsilon}\ +\ Ce^{-m\varepsilon}\ $$
$$u^,(-\varepsilon)\ =\ kA_1cosk(a\ -\ \varepsilon)\ =\ m(-Be^{m\varepsilon}\ +\ Ce^{-m\varepsilon})$$
$$u(+\varepsilon)\ =\ A_2sink(a\ -\ \varepsilon)\ =\ Be^{-m\varepsilon}\ +\ Ce^{m\varepsilon}$$
$$u^,(+\varepsilon)\ =\ kA_2cosk(a\ -\ \varepsilon)\ =\ m(Be^{-m\varepsilon}\ +\ Ce^{m\varepsilon})$$


После исключения $A_1$ и $A_2$ из равенств выше, соответственно остаются два соотношения:

$$kcotk(a\ -\ \varepsilon)\ =\ m\frac{-Be^{2m\varepsilon}+C}{Be^{2m\varepsilon}\ +\ C}$$
$$kcotk(a\ -\ \varepsilon)\ =\ m\frac{B\ -\ Ce^{2m\varepsilon}}{B\ +\ Ce^{2m\varepsilon}}$$


Из тождественности их левых частей следует равенство правых частей. Последние же равны в том и только в том случае, если $B=±C$. При $B=+C$ из равенств требующих непрерывность следует, что $A_1$ = $−A_2$, и мы получаем решения с положительной четностью, если же $B=C$, то $A_1$ = $A_2$, и мы получаем решения с отрицательной четностью. Теперь перейдем к пределу $\varepsilon → 0, \varkappa → ∞$ так, чтобы $\varkappa\varepsilon\ → 0$, но величина $\varkappa^2\varepsilon\ =\ \Omega\ $   оставалась конечной. Величину $Ω$ будем называть коэффициентом непроницаемости перегородки, так как с ростом $Ω$ она становится все более непроницаемой. При положительной четности $(B=C)$ соотношение после разложения его правой части в ряд дает $kcotka\ =\ -\Omega$. Для нечетных получаем $kcotka\ = −∞$ Второй случай проще. Собственные функции обращаются на перегородке в нуль, так что решение имеет вид

$$u_n^-(x)=Asink_n^-(x+a),\ -a \le x<0;\ Asink_n^-(x\ -\ a),\ 0<x \le a$$

Где $$k_n^-a\ =\ \pi n,\ n\ =\ 1,2,3...,\ u_n^-(-x)={-u}_n^-(x),\ {\ |A|}^2\ =\ \frac{1}{a}$$

```python
# Построим график осцилляторного потенциала
def pot(omega, X):
    return 0.5 * constants.electron_mass * pow(omega, 2) * pow(X, 2)

x = np.arange(-3e-2, 3e-2, 1e-4)
y = []
omega = 1
alpha = constants.electron_mass * omega / constants.hbar

for index in range(len(x)):
    y.append(pot(omega, x[index]))

plt.plot(x, y, linewidth=3)

# График нечетных собственных функций для случая когда в точке x = 0
#  вводится перегородка.
def oscillographFunctions2(x, E):
    points = []
    for index in range(len(x)):
        norm = 6e-40  # Коэффициент нормализации,
        #чтобы графики получились не огромные
        if (x[index] < 0):
            k = np.sqrt(2 * m * E) / h
            point = 1/np.sqrt(a) * np.sin(k * (x[index] + a)) * norm + E
            points.append(point)
        else:
            k = np.sqrt(2*m*E)/h
            point = 1/np.sqrt(a) * np.sin(k * (x[index] - a)) * norm + E
            points.append(point)

    return points


# Нарисуем линии энергетических уровней
# Собственные значения — значения энергий
ownValues = []

n = 0
while True:
    E = (n + 0.5) * constants.hbar * omega

    if (E > y[0]):
        break

    ownValues.append(E)
    plt.plot(x, [E] * len(x), linewidth=0.6, c='red')
    plt.plot(x, oscillographFunctions2(x, E), c='green', linewidth=1)
    n += 1

# Рисуем вертикальную перегородку при x = 0
plt.axvline(x=0, color='red', linestyle='--')

# Общие оси и наименования графиков
plt.xlabel("x", fontsize=14)
plt.ylabel("Ψ", rotation=0, fontsize=14)
plt.grid()

# Отображение графиков
plt.show()
```

Данные функции обращаются в ноль у стенок и перегородки для обеспечения непрерывности и нечетности.

В случае непроницаемой перегородки Ω → ∞ она при х=0 обращается в нуль, и обе части ящика становятся независимыми друг от друга. За исключением знака в левой части ящика эта функция совпадает с первой нечетной собственной функцией, показанной в нижней части Графика 2. Таким образом, в пределе Ω → ∞ энергетические уровни становятся вырожденными (крайняя правая полоса на Графике 1).

### Выводы и анализ результатов работы.
Были найдены связные состояния и собственные значения. Также были построены модели и графики с помощью языка Python. Выводы, полученные из работы:
• В яме любой ширины и глубины есть по крайней мере один четный дискретный уровень.

• При любых конечных значениях $а$ и $U$ число $N$ дискретных уровней энергии $E_n$ в яме конечно и $−U < E_n < 0, n = 1, . . . , N$

• Чем больше глубина ямы $U$, и чем больше ее ширина $a$, тем больше число дискретных уровней энергии в яме

• Наименьшей энергией $E_1$(энергией основного состояния) обладает четное состояние

• Значения энергий четных и нечетных состояний чередуются, т. е. уровень энергии нечетного состояния следует за уровнем энергии четного состояния, и наоборот

• Число узлов волновой функции, соответствующей энергии с номером $n (n=1, 2…,N)$, равно $n-1$
